```python
from py2cytoscape import cyrest
from statistics import *
```

```python
cy = cyrest.cyclient()
data = []

raw_data = cy.node.get_properties(network='CURRENT', nodeList='all', propertyList='Y Location, Height')
for entry in raw_data:
    data.append ({
        'suid': str(entry['SUID']),
        'top': entry['visualProperties'][0]['value'] + entry['visualProperties'][1]['value'] / 2,
        'bottom': entry['visualProperties'][0]['value'] - entry['visualProperties'][1]['value'] / 2
    })
    
data
```

```python
def key_function(item):
    return item['bottom']

sorted_data = sorted(data, key=key_function)
```
